```
import math
from ast import literal_eval
def merc(Coords):
    Coordinates = literal_eval(Coords)
    
    lat = Coordinates[0]
    lon = Coordinates[1]

    r_major = 6378137.000
    
    x = r_major * math.radians(lon)
    scale = x/lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + lat * (math.pi/180.0)/2.0)) * scale
    
    return (x, y)
```
